### **String Slicing**

```python
a = 'programming'
```

String Slicing:

Accessing a part / portion of a string by referring to range of indices.

Syntax:
  
    string [ start_index : end_index]
  
  note: end_index is not included.

  

```python
print(a[2:4])
```

```python
print(a[2:]) # it will take the entire string after the index 2
print(a[:9]) # it will start from the 0th index
print(a[:]) # it will take the entire string
```

```python
#slicing with negative index
print(a[-8 : -1])
print(a[-4:])
```

#### Step sizing:


**start_index:** The index where the slice begins (inclusive). If omitted, it defaults to the beginning of the string.
**end_index:** The index where the slice ends (exclusive). If omitted, it defaults to the end of the string.
**step_size:** The number of characters to skip between each character in the slice. If omitted, it defaults to 1.
In the selected code:

- **print(a[0:11:1]):** This slices the string a from index 0 to index 11 (exclusive) with a step size of 1. This effectively prints the entire string programming as the step size of 1 includes every character.

- **print(a[0:11:2]):** This slices the string a from index 0 to index 11 (exclusive) with a step size of 2. This prints every second character starting from the first character, resulting in pormig.


```python
#step size :

print(a[0:11:1])
print(a[0:11:2])
print(a[0:11:3])
print(a[0:11:4])
```

```python
print(a[0:7:3])
```
