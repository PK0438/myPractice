## **Set**

- Set is a collection of heterogeneous data items, which are unordered (no index).
    - not sure about the order how they stored
- Set items are enclosed with in {}
- Set items are immutable
- We can add/remove items to/from a set
- Set doesn't allow duplicate values


```python
S = {23, 56, 3+8j, "python", True, 56}
```

```python
S
```

```python
# Creating an empty set
```

## **Dictionary**

- Dictionary is a collection of Data Items, each item is a key:value pair
- Dictionary items are written with in the  {}
- Dictionary items can be accessed by their key names
- Key duplication is not allowed

```python
D = { 'Brand': 'Hyndai',
     'Model' : 'Tuscon',
      'YOM' : 2025,
      'Color' : 'Rockwood green',
      'Speed' : 320}
```

```python
D
```

```python
## adding a new item


D['Fuel Type'] = 'Gasolin'
```

```python
D
```

```python
## if you use a same key to add, it will update the existing.

D['Color'] = 'Grey'
```

```python
D
```
