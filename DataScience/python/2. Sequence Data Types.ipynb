### **Sequence Data Types**


1. **List**       - collection of heterogeneous data items and enclosed by [], mutable and indexed
2. **Tuple**      - holds any kinds of data and enclosed by (), immutable
3. **Set**        - holds any kinds of data, prints no repetetion, enclosed by {}
4. **Dictionary** - holds the data in key, value pairs, enclosed by {}
5. **String**





#### 1. **List**

```python
a = [1, 2, 3, 4, 5]
print(a)
```

```python
a = [1, 2, 3, 4, "numbers"]
print(a)
```

```python
# Adding items to the list

cities = ["Mumbai", "Delhi", "Bangalore"]
print(cities)

cities.append("Chennai")
print(cities)
```

```python
# Adding multiple items to the list at the sametime.

cities.extend(["Kolkata", "Delhi"])
print(cities)
```

```python
# String cancatination

usCities = ["New york", "Atlanta", "California"]

totalCities = cities + usCities
print(totalCities)
```

```python
# inserting the item at the required index

cities.insert(3, "Hyderabad")
print(cities)
```

#### **Removing** -  removing an item from the list

```python
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata', 'Delhi', 'New york', 'Atlanta', 'California']
cities.remove("Mumbai")
print(cities)
```

#### **del**   -  deleting multiple items in the list, uses index

```python
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata', 'Delhi', 'New york', 'Atlanta', 'California']
del cities[2:5]
print(cities)
```

#### **pop** - it takes the index to pop out the item from the list. if no index specified it will remove from the last index item

```python
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata', 'Delhi', 'New york', 'Atlanta', 'California']
cities.pop()
print(cities)
```

```python
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata', 'Delhi', 'New york', 'Atlanta', 'California']
cities.pop(4)
print(cities)
```
